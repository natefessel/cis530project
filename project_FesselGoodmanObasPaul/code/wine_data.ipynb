```python
from collections import defaultdict
def ngram_counts(filename, n):
    grams = defaultdict(int)
    with open(filename, 'r') as f:
        for line in f:
            words = line.strip().split()
            for i in range(len(words) - n):
                grams[tuple(words[i:i + n])] += 1
    return grams
```

```python
def calculate_overlap(other_bigrams, train_bigrams):
    overlap = 0
    total = 0
    for bigram, count in other_bigrams.items():
        if bigram in train_bigrams:
            overlap += count
        total += count
    return overlap / total, overlap, total
```

```python
for i in range(1, 9):
    train_bigrams = ngram_counts(path.join(wine_reviews_dir, 'wine_reviews_train.txt'), i)
    human_bigrams = ngram_counts(path.join(working_dir, 'mturk', 'human.txt'), i)
    markov_bigrams = ngram_counts(path.join(working_dir, 'mturk', 'markov.txt'), i)

    markov_overlap = calculate_overlap(markov_bigrams, train_bigrams)
    human_overlap = calculate_overlap(human_bigrams, train_bigrams)
    print(r'{} & {:.1f}\% & {:.1f}\% \\ \hline'.format(i, 100 * markov_overlap[0], 100 * human_overlap[0]))
```
