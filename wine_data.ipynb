```python
import pandas as pd
import os.path as path
import os
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
working_dir = os.getcwd()
wine_reviews_dir = path.join(working_dir, 'data', 'wine-reviews')
```

```python
wine_reviews_pt1 = pd.read_csv(path.join(wine_reviews_dir, 'winemag-data_first150k.csv'))
wine_reviews_pt2 = pd.read_csv(
    path.join(wine_reviews_dir, 'winemag-data-130k-v2.csv')).drop(columns=['taster_name', 'taster_twitter_handle', 'title'])
wine_reviews_df = wine_reviews_pt1.append(wine_reviews_pt2).drop(columns=['Unnamed: 0', 'region_1', 'region_2']).drop_duplicates()
print('num samples:', len(wine_reviews_df))
```

```python
X_train, X_test = train_test_split(wine_reviews_df, test_size=1/10)
X_train, X_val = train_test_split(X_train, test_size=1/9)
```

```python
print('training samples: {}, validation samples: {}, test samples: {}'.format(len(X_train), len(X_val), len(X_test)))
```

```python
X_train.to_csv(path.join(wine_reviews_dir, 'wine_reviews_train.csv'), index=False)
X_val.to_csv(path.join(wine_reviews_dir, 'wine_reviews_val.csv'), index=False)
X_test.to_csv(path.join(wine_reviews_dir, 'wine_reviews_test.csv'), index=False)



# for col in wine_reviews_df.columns:
#     print(col)
#     display(wine_reviews_df[col].value_counts()[:5])
```
